```python
import matplotlib.pyplot as plt


classes = ["Python", "R", "AI", "ML", "DS"]
# out of 100 student in each class
class1_students = [30, 10, 20, 25, 10] 
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]


class3_students
```

```python
plt.bar(classes,class1_students)
```

```python
plt.barh(classes,class1_students)
```

```python
import numpy as np

classes = ["Python", "R", "AI", "ML", "DS"]

x_val=np.arange(len(classes))

plt.xticks(x_val,classes)

plt.bar(x_val,class1_students)

```

```python
x_val=np.arange(len(classes))

plt.xticks(x_val,classes)

plt.bar(x_val+0.2, class1_students , label="class1_students", width=0.2)
plt.bar(x_val-0.2, class2_students , label="class2_students", width=0.2)
plt.bar(x_val    , class3_students , label="class3_students", width=0.2)

plt.legend()
plt.show()
```

```python
y_val=np.arange(len(classes))

plt.yticks(x_val,classes)

plt.barh(x_val+0.2, class1_students , label="class1_students", height=0.2)
plt.barh(x_val-0.2, class2_students , label="class2_students", height=0.2)
plt.barh(x_val    , class3_students , label="class3_students", height=0.2)

plt.legend()
plt.show()
```

```python
x_val=np.arange(len(classes))

plt.xticks(x_val,classes)

plt.title("Student selected Languages")
plt.xlabel("Selected Languages")
plt.ylabel("No of Students")

plt.bar(x_val+0.2, class1_students , label="class1_students", width=0.2)
plt.bar(x_val-0.2, class2_students , label="class2_students", width=0.2)
plt.bar(x_val    , class3_students , label="class3_students", width=0.2)

plt.legend()
plt.show()
```
