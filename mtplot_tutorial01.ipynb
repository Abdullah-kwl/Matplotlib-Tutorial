# Intro to Matplotlib

```python
import matplotlib.pyplot as plt

# 'x' are the days
# 'y' is the respective temperature

x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]

plt.plot(x,y)
```

Now we will see which arguments plot() function take

```python
plt.plot(x,y, color='green')
```

```python
plt.plot(x,y, color='green' ,linewidth=5)
```

```python
plt.plot(x,y, color='green' ,linewidth=3 , linestyle='dotted' )
```

```python
# see the documentaion
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot(x,y, color='g' ,linewidth=3 , linestyle='-.' ,marker="o")
```

# Now we will add some labels

```python
plt.title("Weather")
plt.xlabel("Days")
plt.ylabel("Temperature")

plt.plot(x,y, color='g' ,linewidth=3 , linestyle='-.' ,marker="o")
```

```python
# code copy from w3 school web
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}


plt.title("Weather" , fontdict = font1)
plt.xlabel("Days" , fontdict = font2)
plt.ylabel("Temperature" , fontdict = font2)

plt.plot(x,y, color='g' ,linewidth=3 , linestyle='-.' ,marker="o")
```
