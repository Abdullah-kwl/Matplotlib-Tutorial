# Formate string use in plot() fun in Detail
see the documentaion at:
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
import matplotlib.pyplot as plt

# 'x' are the days
# 'y' is the respective temperature

x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]

plt.plot(x,y)
```

```python
# we will use F.M.T(formate dtring)
plt.plot(x,y, 'o-.r' )
```

```python
# fmt = '[marker][line][color]'
plt.plot(x,y, marker='o',linestyle='-.',color='red')
```

```python
plt.plot(x,y, marker='*',linestyle='-.',color='r')
```

```python
# ,markersize=15
plt.plot(x,y, marker='*',linestyle='-.',color='r',markersize=15)
```

```python
plt.plot(x,y, marker='o',markersize=15,linestyle=':',linewidth=5,color='r')
```

```python
plt.plot(x,y, marker='o',markersize=15,linestyle='-',linewidth=5,color='r')
```

Alpha is the opecity (darkness value between 0-1 , 0 to 1)

```python
plt.plot(x,y, marker='o',markersize=15,linestyle='-',linewidth=5,color='r', alpha=0.1)
```

```python
plt.plot(x,y, marker='o',markersize=15,linestyle='-',linewidth=5,color='r', alpha=0.3)
```

```python
plt.plot(x,y, marker='o',markersize=15,linestyle='-',linewidth=5,color='r', alpha=0.5)
```

```python
plt.plot(x,y, marker='o',markersize=15,linestyle='-',linewidth=5,color='r', alpha=0.8)
```

```python
plt.plot(x,y, marker='o',markersize=15,linestyle='-',linewidth=5,color='r', alpha=1)
```

Now we will see some property of labels

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]

plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
# now we will specify which line is for which lable
plt.plot(days, max_t , label="Max temperature")
plt.plot(days, min_t , label="Min temperature")
plt.plot(days, avg_t , label="Avg temperature")
```

```python
# now we will specify which line is for which lable
# We will have to use 'legend' to show lables
plt.plot(days, max_t , label="Max temperature" )
plt.plot(days, min_t , label="Min temperature")
plt.plot(days, avg_t , label="Avg temperature")

plt.legend()
```

```python
plt.plot(days, max_t , label="Max temperature" )
plt.plot(days, min_t , label="Min temperature")
plt.plot(days, avg_t , label="Avg temperature")

plt.legend()
plt.grid()
```

```python
plt.plot(days, max_t , label="Max temperature" )
plt.plot(days, min_t , label="Min temperature")
plt.plot(days, avg_t , label="Avg temperature")

plt.legend(loc='best', fontsize="large",shadow=True)
plt.grid()
```
