# Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("googleplaystore.csv")
df
```

```python
df2=df[df["Rating"]<10.0]

# x=df["Rating"]

x=df2["Rating"]
y=df2["Reviews"]
```

```python
plt.figure(figsize=(16 , 9))

plt.title("Play Store Data")
plt.xlabel("Rating")
plt.ylabel("Reviews")

plt.scatter(x,y)
```
