```python
import matplotlib.pyplot as plt

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

plt.plot(company,revenue, label="revinue")
plt.plot(company,profit,  label="profit")
plt.legend(fontsize=15 ,shadow=True)
```

# Why we use the xpose array inted of company?

```python
# now changing the index of x-axes with company names
import numpy as np

# we will take a dummy array and put it insted of company name
xpos=np.arange(len(company))

# we will use the 'xticks' funtion to set the company name at x-axes
plt.xticks(xpos,company)

# adding some labes
plt.title("Us Companies Stock" , color="#0467b3" , fontsize=20)
plt.ylabel("revenue(bln)"      , color="#fc8403" , fontsize=20)

# we can set the width of bar using 'width' 
# we will add and subtract from xpose therefor we use the numaric values insted of company names
plt.bar(xpos-0.2,revenue, width=0.4 ,label='Revenue')
plt.bar(xpos+0.2,profit, width=0.4  ,label='Profit')

plt.legend(fontsize=10 , shadow=True)
```

# Horizontal bar chart

```python
# now changing the index of x-axes with company names
import numpy as np

# we will take a dummy array and put it insted of company name
xpos=np.arange(len(company))

# we will use the 'yticks' funtion to set the company name at y-axes
plt.yticks(xpos,company)

# adding some labes
plt.title("Us Companies Stock" , color="#0467b3" , fontsize=20)
plt.xlabel("revenue(bln)"      , color="#fc8403" , fontsize=20)

# we can set the height of horizental-bar insted of width 
# we will add and subtract from xpose therefor we use the numaric values insted of company names
plt.barh(xpos-0.2,revenue, height=0.4  ,label='Revenue')
plt.barh(xpos+0.2,profit,  height=0.4   ,label='Profit')

plt.legend(fontsize=10 , shadow=True)
```
